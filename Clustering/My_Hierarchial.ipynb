# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot  as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:,[3,4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
from scipy.cluster import  hierarchy as hier
dendogram = hier.dendrogram(hier.linkage(x, method = 'ward'))
plt.title('Dendogram')
plt.xlabel('Points')
plt.ylabel('Ecludian Distance')
plt.axhline(350, linestyle='dashed', color='k')
plt.annotate(xy=[1000, 350], s='{} Clusters'.format(3))
plt.axhline(200, linestyle='dashed', color='k')
plt.annotate(xy=[1000, 200], s='{} Clusters'.format(5))
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
clusters = AgglomerativeClustering(n_clusters=3)
y_pred = clusters.fit_predict(x)
```

```python
y_pred
```

## Visualising the clusters

```python
plt.scatter(x[y_pred ==0, 0], x[y_pred == 0, 1], s = 100, c ='red', label='Cluster 1')
plt.scatter(x[y_pred ==1, 0], x[y_pred == 1, 1], s = 100, c ='blue', label='Cluster 2')
plt.scatter(x[y_pred ==2, 0], x[y_pred == 2, 1], s = 100, c ='green', label='Cluster 3')
# plt.scatter(x[y_pred ==3, 0], x[y_pred == 3, 1], s = 100, c ='yellow', label='Cluster 4')
# plt.scatter(x[y_pred ==4, 0], x[y_pred == 4, 1], s = 100, c ='black', label='Cluster 5')

plt.title('This is Clustering')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')

plt.legend()
plt.show()
```
